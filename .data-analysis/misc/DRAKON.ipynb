```python
import pandas as pd                             # imports

path = "./CMO-Historical-Data-Annual.xlsx"      # path to dataset

data = pd.read_excel(path,sheet_name=1)         # initalizing the data
```

```python
name_row = data.iloc[5:6]
attribute_row = data.iloc[6:7]

name_row

some_list = []

for i in name_row:
    column = f"""{name_row[i].values[0]} {attribute_row[i].values[0]}"""
    some_list.append(column)

some_list[0] = "Year"


df = data
df.columns = some_list
df = df.iloc[7:]
df.set_index("Year", inplace=True)
```
