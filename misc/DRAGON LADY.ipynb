```python
import pandas as pd                             # imports

path = "./CMO-Historical-Data-Annual.xlsx"      # path to dataset

data = pd.read_excel(path,sheet_name=1)         # initalizing the data
```

```python
name_row = data.iloc[5:6]
attribute_row = data.iloc[6:7]

name_row

some_list = []

for i in name_row:
    # print(f"""{name_row[i].values[0]} : {attribute_row[i].values[0]} """)
    column = f"""{name_row[i].values[0]} {attribute_row[i].values[0]}"""
    some_list.append(column)

# print(some_list)
some_list[0] = "Year"


for i in range(10):
    print(some_list[i])
```

```python
# COLUMN_NAMES = [

#     # Crude oil
#     "Crude Oil (avg)",
#     "Crude Oil (Brent)",
#     "Crude Oil (Dubai)",
#     "Crude Oil (WTI)",


#     #Coal
#     "Coal (AU)",
#     "Coal (ZA)",
    

#     #Natural Gas
#     "Natural Gas (US)",
#     "Natural Gas (EU)",
#     "Liquefied Natural Gas",


#     #Metal
#     "Aluminium", "Iron Ore (CFR spot)", "Copper", "Lead", "Tin", "Nickel", "Zinc", "Gold", "Platinu", "Silver"
# ]

df = data

df.columns = some_list
df = df.iloc[7:]
df.set_index("Year", inplace=True)
df
```

```python
CRUDE_OIL_AVG = df["Crude oil, average ($/bbl)"]
CRUDE_OIL_WTI = df['Crude oil, WTI ($/bbl)']
```

```python
# Check for missing values (including replaced NaNs)
missing_values = df['Crude oil, WTI ($/bbl)'].isna()

# Convert non-missing values to numeric, replacing errors with np.nan
CRUDE_OIL_WTI.loc[~missing_values] = pd.to_numeric(CRUDE_OIL_WTI[~missing_values], errors='coerce')
CRUDE_OIL_WTI
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,17))

# creating figure (1x1)
fig , ax = plt.subplots(figsize=(14,5))   


# plotting data
ax.plot(df['Crude oil, average ($/bbl)'],   label="Crude Oil Average")
ax.plot(df['Crude oil, Brent ($/bbl)']  ,   label="Crude oil Brent")
ax.plot(df['Crude oil, Dubai ($/bbl)']  ,   label="Crude oil Dubai")
# ax.plot(df['Crude oil, WTI ($/bbl)']    ,   label="Crude oil WTI")


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.0, 5.0, 100)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)

# Create a figure with 1 row and 1 column
fig, ax = plt.subplots()  # 1 row, 1 column

# Plot data on the same axes, using different line styles
ax.plot(x, y1, label='Sine', linestyle='-')  # Solid line
ax.plot(x, y2, label='Cosine', linestyle='--')  # Dashed line

# Add labels and title
ax.set(xlabel='X', ylabel='Y')
fig.suptitle('Sine and Cosine Functions')

# Add legend
ax.legend(df[''])

plt.show()
```
